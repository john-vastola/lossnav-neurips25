**Caption.** (a) Example optimal $\theta_t$ traces for a 1D quadratic loss, assuming different values of the temporal discounting rate ($\gamma = 0, 1, 10$). (b) Loss over time given the $\theta_t$ from (a), same values of $\gamma$. Note that lower values of $\gamma$ (`longer' planning horizon) produce loss curves that converge more quickly. (c) Shape of trajectories for different $\gamma$ given a 2D anisotropic loss. In the gradient-descent-like regime ($\gamma \gg 1$), $\theta_2$ converges much more quickly than $\theta_1$ due to the anisotropy. In the ballistic ($\gamma \approx 0$) regime, the difference in convergence rates is not as extreme. (d) Ratio of convergence rates $r_i := \sqrt{\gamma^2/4 + \eta k H_{ii}} - \gamma/2$ assuming a diagonal Hessian. In the gradient-descent-like regime, directions with four times as much curvature converge $4$ times faster; in the ballistic regime, they only converge $\sqrt{4} = 2$ times faster. (e) An Adam-like implementation of the ballistic ($\gamma \approx 0$) rule was used to train a small multilayer perceptron (MLP) to classify MNIST digits. Left: loss over training, right: test set accuracy over training. (f) Same as in (e), but for a small convolutional neural network (CNN) trained to classify CIFAR-10 images. The ballistic rule generally performs better than SGD (black), and similarly to or worse than Adam (red). 

## a, b. Optimal trajectory and loss curve for different values of temporal discounting rate

```python
def get_curves_qloss(t, params):
    gamma, k, eta = params['gamma'], params['k'], params['eta']
    h = params['h']
    mu = params['mu']
    theta_init = params['theta_init']

    
    r =  np.sqrt( (gamma/2)**2 + eta*k*h ) - (gamma/2)   # shape: D
    theta = (1 - np.exp(-r[None,:]*t[:,None]))*mu[None,:] +  np.exp(-r[None,:]*t[:,None])*theta_init[None,:]  # shape: num_t, D

    L = np.sum( 0.5*h[None,:]*((theta - mu[None,:])**2), axis=-1) # shape: num_t

    J_prefactor = (1/(2*eta))*( r**2 + eta*k*h )*((theta_init - mu)**2)/(2*r + gamma)
    J = J_prefactor[None,:]*( 1 - np.exp(- (2*r[None,:] + gamma)*t[:,None]))
    return theta, L, J
```

```python
t_min, t_max, num_t = 0, 10, 2000
t = np.linspace(t_min, t_max, num_t)

gamma = 10
k = 1
eta = 1

h = np.array([1])
mu = np.array([1])
theta_init = np.array([0])

params = {'gamma':gamma, 'k':k, 'eta':eta, 'h':h, 'mu':mu, 'theta_init':theta_init}
```

```python
params['gamma'] = 0.0
theta_low, L_low, J_low = get_curves_qloss(t, params)

params['gamma'] = 1
theta_med, L_med, J_med = get_curves_qloss(t, params)

params['gamma'] = 10
theta_hi, L_hi, J_hi = get_curves_qloss(t, params)
```

```python
fig, ax = plt.subplots(figsize=(5,5))
plt.plot(t, theta_low, linewidth=lw, color=color_low)
plt.plot(t, theta_med, linewidth=lw, color=color_med)
plt.plot(t, theta_hi, linewidth=lw, color=color_hi)
ax.spines[['right', 'top']].set_visible(False)
plt.xticks([0, 5, 10], fontsize=25)
plt.yticks([0.5, 1], ['0.5','1'], fontsize=25)
plt.xlim(0, None); plt.ylim(0, None)

save_plot('fig2_mom_1D_theta')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5,5))
plt.plot(t, L_low, linewidth=lw, color=color_low)
plt.plot(t, L_med, linewidth=lw, color=color_med)
plt.plot(t, L_hi, linewidth=lw, color=color_hi)

ax.spines[['right', 'top']].set_visible(False)
plt.xticks([0, 5, 10], fontsize=25)
plt.yticks([0.25, 0.5], ['0.25','0.5'], fontsize=25)
plt.xlim(0, None); plt.ylim(-0.01, None)

save_plot('fig2_mom_1D_loss')
plt.show()
```

## d. Ratio of convergence rates

```python
def get_t_ratio(h, gamma, params):
    k, eta, h0 = params['k'], params['eta'], params['h0']

    numerator = np.sqrt( (gamma/2)**2 + eta*k*h  ) - gamma/2
    denominator = np.sqrt( (gamma/2)**2 + eta*k*h0  ) - gamma/2

    t_ratio = numerator/denominator
    return t_ratio
```

```python
h_min, h_max, num_h = 0.01, 4, 100
h = np.linspace(h_min, h_max, num_h)


eta = 1
k = 1
gamma = 0
h0 = 1

params = {'gamma':gamma, 'k':k, 'eta':eta, 'h0':h0}
```

```python
t_ratio_low = get_t_ratio(h, 0, params)
t_ratio_med = get_t_ratio(h, 1, params)
t_ratio_hi = get_t_ratio(h, 10, params)
```

## e. Comparison of ballistic rule, SGD, and Adam on MNIST image classification

```python
# -----------------------------
# Repro & small utilities
# -----------------------------
def set_seed(seed: int):
    random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

@dataclass
class RunConfig:
    dataset: str
    model: str
    opt: str
    lr: float
    weight_decay: float
    batch_size: int
    epochs: int
    device: str
    mom: float
    beta2: float
    eps: float
    bleed: float  # velocity bleed for ballistic (0 = perfectly frictionless)
```

```python
# -----------------------------
# Data
# -----------------------------
def get_loaders(dataset: str, batch_size: int) -> Tuple[DataLoader, DataLoader]:
    if dataset.upper() == 'MNIST':
        transform = transforms.Compose([
            transforms.ToTensor(),
            transforms.Normalize((0.1307,), (0.3081,)),   # data standardization
        ])
        train = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
        test = datasets.MNIST(root='./data', train=False, download=True, transform=transform)
    elif dataset.upper() == 'CIFAR10':
        transform_train = transforms.Compose([
            transforms.RandomCrop(32, padding=4),
            transforms.RandomHorizontalFlip(),           # data augmentation, standardization
            transforms.ToTensor(),
            transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2470, 0.2435, 0.2616)),
        ])
        transform_test = transforms.Compose([
            transforms.ToTensor(),
            transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2470, 0.2435, 0.2616)),
        ])
        train = datasets.CIFAR10(root='./data', train=True, download=True, transform=transform_train)
        test = datasets.CIFAR10(root='./data', train=False, download=True, transform=transform_test)
    else:
        raise ValueError('Unsupported dataset')

    train_loader = DataLoader(train, batch_size=batch_size, shuffle=True, num_workers=2, pin_memory=True)
    test_loader = DataLoader(test, batch_size=512, shuffle=False, num_workers=2, pin_memory=True)
    return train_loader, test_loader
```

```python
# -----------------------------
# Models
# -----------------------------
class MLP(nn.Module):
    def __init__(self, num_classes=10):
        super().__init__()
        self.net = nn.Sequential(
            nn.Flatten(),
            nn.Linear(28 * 28, 512), nn.ReLU(),
            nn.Linear(512, 256), nn.ReLU(),
            nn.Linear(256, num_classes),
        )
    def forward(self, x):
        return self.net(x)

class SmallCIFARCNN(nn.Module):
    def __init__(self, num_classes=10):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(3, 64, 3, padding=1), nn.ReLU(),
            nn.Conv2d(64, 64, 3, padding=1), nn.ReLU(),
            nn.MaxPool2d(2),  # 16x16
            nn.Conv2d(64, 128, 3, padding=1), nn.ReLU(),
            nn.Conv2d(128, 128, 3, padding=1), nn.ReLU(),
            nn.MaxPool2d(2),  # 8x8
        )
        self.head = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128 * 8 * 8, 256), nn.ReLU(),
            nn.Linear(256, num_classes),
        )
    def forward(self, x):
        return self.head(self.conv(x))


def make_model(name: str, dataset: str) -> nn.Module:
    if dataset.upper() == 'MNIST':
        num_classes = 10
        if name.lower() == 'mlp':
            return MLP(num_classes)
        else:
            raise ValueError('For MNIST use --model mlp')
    else:
        num_classes = 10
        if name.lower() == 'smallcnn':
            return SmallCIFARCNN(num_classes)
        else:
            raise ValueError('For CIFAR10 use --model smallcnn')
```

```python
# -----------------------------
# Optimizers
# -----------------------------
class Ballistic(optim.Optimizer):

    def __init__(self, params: Iterable[torch.nn.Parameter], lr=1e-2, beta2=0.999, eps=1e-8, weight_decay=0.0, bleed=0.0):
        defaults = dict(lr=lr, beta2=beta2, eps=eps, weight_decay=weight_decay, bleed=bleed)
        super().__init__(params, defaults)

    @torch.no_grad()
    def step(self):
        loss = None
        for group in self.param_groups:
            lr = group['lr']
            beta2 = group['beta2']
            eps = group['eps']
            wd = group['weight_decay']
            bleed = group['bleed']
            for p in group['params']:
                if p.grad is None:
                    continue
                g = p.grad
                state = self.state[p]
                # state init
                if len(state) == 0:
                    state['step'] = 0
                    state['s2'] = torch.zeros_like(p)
                state['step'] += 1
                s2 = state['s2']
                # decoupled weight decay (like AdamW): shrink params directly
                if wd != 0.0:
                    p.add_(p, alpha=-lr * wd)
                # second-moment update (uncentered)
                s2.mul_(beta2).addcmul_(g, g, value=(1.0 - beta2))
                pre = g / (s2.add(eps).sqrt())
                # integrate velocity (frictionless except optional small bleed)
                p.add_(pre, alpha=-lr)
        return loss

# Convenience wrappers for SGD & momentum (for symmetry) -----------------
class SGDOverdamped(optim.SGD):
    def __init__(self, params, lr=0.1, weight_decay=0.0):
        super().__init__(params, lr=lr, momentum=0.0, weight_decay=weight_decay)

class SGDMomentum(optim.SGD):
    def __init__(self, params, lr=0.1, momentum=0.9, weight_decay=0.0):
        super().__init__(params, lr=lr, momentum=momentum, nesterov=False, weight_decay=weight_decay)
        

def make_opt(name: str, params: Iterable[nn.Parameter], cfg: RunConfig):
    name = name.lower()
    if name == 'sgd':
        return SGDOverdamped(params, lr=cfg.lr, weight_decay=cfg.weight_decay)
    if name == 'mom':
        return SGDMomentum(params, lr=cfg.lr, momentum=cfg.mom, weight_decay=cfg.weight_decay)
    if name == 'adam':
        return optim.AdamW(params, lr=cfg.lr if cfg.lr > 0 else 1e-3, weight_decay=cfg.weight_decay, )
    if name == 'ballistic':
        return Ballistic(params, lr=cfg.lr, beta2=cfg.beta2, eps=cfg.eps, weight_decay=cfg.weight_decay, bleed=cfg.bleed)
    raise ValueError('Unknown optimizer')
```

```python
# TRAINING

def accuracy(model: nn.Module, loader: DataLoader, device: str) -> float:
    model.eval()
    correct = total = 0
    with torch.no_grad():
        for x, y in loader:
            x, y = x.to(device), y.to(device)
            logits = model(x)
            pred = logits.argmax(dim=1)
            correct += (pred == y).sum().item()
            total += y.numel()
    return 100.0 * correct / total


def train_epoch(model, loader, opt, device):
    model.train()
    total_loss = 0.0
    for x, y in loader:
        x, y = x.to(device), y.to(device)
        opt.zero_grad(set_to_none=True)
        logits = model(x)
        loss = F.cross_entropy(logits, y)
        loss.backward()
        # simple global grad clip initial steps
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=5.0)
        opt.step()
        total_loss += loss.item() * y.size(0)
    return total_loss / len(loader.dataset)


def train_me(model, train_loader, test_loader, opt, cfg):
    
    best = 0.0
    train_loss_list = []
    test_acc_list = []
    for epoch in range(1, cfg.epochs + 1):
            train_loss = train_epoch(model, train_loader, opt, cfg.device); train_loss_list.append(train_loss)
            test_acc = accuracy(model, test_loader, cfg.device); test_acc_list.append(test_acc)
            best = max(best, test_acc)
        
            print(f'Epoch {epoch:03d}  loss {train_loss:.4f}  test {test_acc:.2f}%  (best {best:.2f}%)')
    return model, train_loss_list, test_acc_list
```

```python
dataset = 'MNIST'
model_name = 'mlp'
opt_name = 'ballistic'
lr = 5e-4
epochs = 10
device ='mps' #if torch.cuda.is_available() else 'cpu'    # CHANGE IF NOT USING M1 MAC
batch_size = 128

bleed = 0.
eps = 1e-8
beta2 = 0.999
mom = 0.9
weight_decay = 0.


lr_sgd = 1e-1
lr_adam = 1e-3
lr_ball = 5e-4   # better: 5e-4?

cfg_sgd_m = RunConfig(
    dataset=dataset, model=model_name, opt='sgd', lr=lr_sgd,
    weight_decay=weight_decay, batch_size=batch_size,
    epochs=epochs, device=device, mom=mom,
    beta2=beta2, eps=eps, bleed=bleed,
)

cfg_adam_m = RunConfig(
    dataset=dataset, model=model_name, opt='adam', lr=lr_adam,
    weight_decay=weight_decay, batch_size=batch_size,
    epochs=epochs, device=device, mom=mom,
    beta2=beta2, eps=eps, bleed=bleed,
)

cfg_ball_m = RunConfig(
    dataset=dataset, model=model_name, opt='ballistic', lr=lr_ball,
    weight_decay=weight_decay, batch_size=batch_size,
    epochs=epochs, device=device, mom=mom,
    beta2=beta2, eps=eps, bleed=bleed,
)
```

### SGD

Set up model.

```python
# Load data, make train and test sets
train_loader_sgd_m, test_loader_sgd_m = get_loaders(cfg_sgd_m.dataset, cfg_sgd_m.batch_size)

# Load model
model_sgd_m = make_model(cfg_sgd_m.model, cfg_sgd_m.dataset).to(cfg_sgd_m.device)

# Load optimizer
opt_sgd_m = make_opt(cfg_sgd_m.opt, model_sgd_m.parameters(), cfg_sgd_m)
```

Train. 

```python
model_sgd_m, train_sgd_m, test_sgd_m = train_me(model_sgd_m, train_loader_sgd_m, test_loader_sgd_m, opt_sgd_m, cfg_sgd_m)
```
